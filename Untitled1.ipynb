```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
x = np.arange(0, 10, 0.1)

f, ax = plt.subplots()
ax.plot(x, np.sin(x));

f.savefig('figure.pdf')
```

```python
f.savefig('figure_300.png', dpi=300) #save as png with resolution adjusted
```

```python
#defining the size of your image inches
f, ax = plt.subplots()

print('Default size:', f.get_size_inches() * 2.54)

f.set_size_inches(10 / 2.54, 5 / 2.54)

ax.plot(x, np.sin(x));

print('Manually set size:', f.get_size_inches() * 2.54)
```

```python
#changing displayed size without changing saved file
default_dpi = mpl.rcParams['figure.dpi']
default_dpi #original
```

```python
mpl.rcParams['figure.dpi'] = 5
f, ax = plt.subplots()
ax.plot(x, np.sin(x));
```
