```python
import matplotlib.pyplot as plt
import netCDF4 as nc
import numpy as np

%matplotlib inline
```

```python
#artificial data
x = np.arange(0, 3 * np.pi, 0.01)

y = np.sin(x)
# name of the data file
fN = '../../data/co2.nc'

ncf = nc.Dataset(fN)

year = ncf.variables['year'][:]
co2 = ncf.variables['co2_annual'][:]

print(year[:3])
print(co2[:3])
```

```python
plt.plot(x,y)
```

```python
plt.plot(year,co2)
```

```python
#better to explicitely create a figure and an axes instance
f,ax=plt.subplots()
#f is the figure - canvas containing everything including axes
#ax is object of class plt.axes. i.e it contains the elements like axes, ticks, labels
ax.plot(x,y)  #calling the plot intance of the axes
```

```python
ax.plot(year,co2) #figure istance needs to be created with
```

```python
f,ax=plt.subplots()
ax.plot(year,co2)
```

```python
#Using plt.subplots() with no arguments is equal to plt.subplots(1, 1)
f,ax=plt.subplots(2,1) #(nrows,ncolumns)
```

```python
#yu can add another line on the same plot like
f, ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
```

```python
f, ax = plt.subplots()

ax.plot(x, y + 0, color='green')            # specify color by name
ax.plot(x, y + 1, color='b')                # color code (rgbcmyk)
ax.plot(x, y + 2, color='0.75')             # Greyscale between 0 and 1
ax.plot(x, y + 3, color='#1c9099')          # Hex code (RRGGBB from 00 to FF)
ax.plot(x, y + 4, color=(1.0,0.2,0.3))      # RGB tuple, values 0 to 1
ax.plot(x, y + 5, color=(1.0,0.2,0.3, 0.5)) # RGB tuple including alpha channel (transparency)
ax.plot(x, y + 6, color='chartreuse');      # all HTML color names supported

ax.plot(x, y + 7, color='none');            # 'none' stands for no color -invisible
```

```python
#observation usually shown in black
f, ax = plt.subplots()

ax.plot(year, co2 + 0, color='black')       # specify color by name
ax.plot(year, co2 + 2, color='k')           # color code (rgbcmyk)
ax.plot(year, co2 + 4, color='0')           # Greyscale between 0 and 1
ax.plot(year, co2 + 6, color='#000000')     # Hex code (RRGGBB from 00 to FF)
ax.plot(year, co2 + 8, color=(0, 0, 0))     # RGB tuple, values 0 to 1
```

```python
#line styles
f, ax = plt.subplots()

ax.plot(x, x + 0, linestyle='solid')
ax.plot(x, x + 1, linestyle='dotted')
# using ls also works
ax.plot(x, x + 2, ls='dashdot')


ax.plot(x, x + 5, linestyle='-') # solid
ax.plot(x, x + 6, linestyle=':') # dotted
# using ls also works
ax.plot(x, x + 7, ls='-.') # dashdot
```

```python
f, ax = plt.subplots()

ax.plot(x, np.sin(x), label='sin(x)')
ax.plot(x, np.cos(x), label='cos(x)')

# no label keyword -> no legend entry
ax.plot(x, x / 10)

ax.set_ylim(-1.1, 1.1)

ax.legend()
```

```python
f, ax = plt.subplots(1, 1)

ax.plot(year, co2, color='0.1',ls='dotted', label='CO2 concentration') #ls short for linestyle

ax.legend()

ax.set_title('Atmospheric CO$_2$ [ppm] manoa loa', fontsize=14)

ax.set_xlabel('time')

ax.set_ylabel('CO$_2$ [ppm]')
```

```python
#latex command
f, ax = plt.subplots()

ax.plot(x, np.cos(x))

ax.set_ylabel(r'$\frac{\mathrm{d} \, \sin(x)}{\mathrm{d} x}$', fontsize=16);

ax.set_xlabel('radian ($^\circ$)', fontsize=12);
ax.set_xlim(2, 6) #set axis limit
ax.set_ylim(1, -1) #min limit larger than minimum value - axis inverted
ax.set_xlim(2, None) #only one way limit
```

```python
#mnaual ticks on graph axis
xticks = np.arange(0, 10, np.pi)

# create latex-formatted tick lables
xticklabels = ['$0$', '$\pi$', '$2 \pi$', '$3 \pi$']

f, ax = plt.subplots()

ax.plot(x, np.cos(x))

ax.set_xticks(xticks);
ax.set_xticklabels(xticklabels);
```
